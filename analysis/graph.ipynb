```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
filepath = "/Users/celine/Documents/uni/UZH/CL/5.Semester/machine-translation/exercises/3/mt-exercise-03/logs"
log_df = []
for logs in os.listdir(filepath):
    log_df.append(pd.read_csv(os.path.join(filepath, logs), sep=","))
log_df = pd.concat(log_df)

```

```python
import pylab as plot
params = {'legend.fontsize': 20,
          'legend.handlelength': 2,
          'patch.linewidth': 2}
plot.rcParams.update(params)
log_df["dropout"] = log_df["dropout"].astype(str)
hue_order = sorted(log_df["dropout"].unique())
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
fig,ax = plt.subplots(1,3,figsize=(16,9),sharex=True, sharey=True)
ax1 = ax[0]
sns.lineplot(data=log_df, x="epoch", y="train-perplexity", hue="dropout", ax=ax1, linewidth=4,hue_order=hue_order)
ax1.set_title("Train Perplexity")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Perplexity")
ax2 = ax[1]
sns.lineplot(data=log_df, x="epoch", y="val-perplexity", hue="dropout", ax=ax2, linewidth=4,hue_order=hue_order)
ax2.set_title("Validation Perplexity")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Perplexity")
ax3 = ax[2]
sns.lineplot(data=log_df, x="epoch", y="test-perplexity", hue="dropout", ax=ax3, linewidth=4,hue_order=hue_order)
ax3.set_title("Test Perplexity")
ax3.set_xlabel("Epoch")
ax3.set_ylabel("Perplexity")
plt.tight_layout()
plt.show()
```
